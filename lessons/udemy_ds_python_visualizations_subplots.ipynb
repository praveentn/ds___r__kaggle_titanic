```python
fg4 = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(s=50, linewidth=0.5, edgecolor='black')
fg4 = fg4.map(plt.scatter, 'CriticRating', 'AudienceRating', **kws)
fg4.set(xlim=(0,100), ylim=(0,100))

for ax in fg4.axes.flat:
    ax.plot((20,60),(20,60), c='gray', ls='--')
fg4.add_legend()
```
