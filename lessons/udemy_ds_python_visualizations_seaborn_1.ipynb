```python
for gen in movies.Genre.cat.categories:
    print(gen)
```

```python
tl = []
for gen in movies.Genre.cat.categories:
    tl.append(movies[movies.Genre == gen].BudgetMillions)
plt.hist(tl, stacked=True)
plt.show()
```

```python
tl = []
for gen in movies.Genre.cat.categories:
    tl.append(movies[movies.Genre == gen].BudgetMillions)
plt.hist(tl, stacked=True, bins=30, rwidth=2)
plt.show()
```

```python
mylabels = []
tl = []
for gen in movies.Genre.cat.categories:
    tl.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)
plt.hist(tl, stacked=True, bins=30, rwidth=2, label=mylabels)
plt.legend()
plt.show()
```
